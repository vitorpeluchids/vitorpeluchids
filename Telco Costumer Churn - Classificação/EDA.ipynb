```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('data\WA_Fn-UseC_-Telco-Customer-Churn.csv', sep =',')
df.head()
```

#ENTENDENDO OS DADOS E LIMPANDO ERROS

```python
#Entendendo os dados
df.describe(), df.dtypes
```

A coluna TotalCharges apresenta alguma inconsistencia fazendo com que ela aponte como "object" e não como int ou float, provavelmente é alguma valor NAN, NUll ou espaço vazio " ".

```python
df.isnull().sum()
```

```python
# Procura por strings que contêm apenas espaços em branco
espacos_vazios = (df.astype(str).apply(lambda x: x.str.isspace() | (x == ""))).sum()
print(espacos_vazios)
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
df.dropna(subset=['TotalCharges'], inplace=True)
```

#EXPLORANDO O DATASET

```python
# Criando o gráfico de barras para a coluna Churn
fig = px.histogram(df, x="Churn", 
                   title='Distribuição da Classe Target (Churn)',
                   color="Churn", # Diferencia as barras por cor
                   text_auto=True, # Adiciona os números em cima das barras
                   color_discrete_sequence=["#636EFA", "#EF553B"]) # Cores customizadas

fig.update_layout(yaxis_title="Quantidade de Clientes", xaxis_title="Churn")
fig.show()
```
